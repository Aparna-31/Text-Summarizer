```python
from transformers import pipeline

# Load the summarization model
summarizer = pipeline("summarization", model="facebook/bart-large-cnn")

# Input text
text = """
Artificial Intelligence is transforming industries by enabling automation,
data-driven insights, and intelligent decision-making. It is used in healthcare,
finance, retail, and many other sectors. With rapid advancements in machine
learning and deep learning, AI systems are expanding their capabilities across
a wide variety of real-world applications.
"""

# Generate summary
summary = summarizer(
    text,
    max_length=60,  # max summary length
    min_length=20,  # min summary length
    do_sample=False
)

print("Summary:")
print(summary[0]['summary_text'])
```

```python
"""
LLM Text Summarizer using HuggingFace Transformers
Model: facebook/bart-large-cnn

What this project does:
- Loads a pre-trained summarization model
- Takes input text from the user
- Generates a concise summary
- Ensures clean pipeline-based implementation
"""

from transformers import pipeline

# Step 1: Load the summarization pipeline
# This downloads the BART model and creates a summarizer function.
summarizer = pipeline("summarization", model="facebook/bart-large-cnn")

def generate_summary(text):
    """
    Function that takes raw text and returns a summary.
    max_length → upper limit of summary
    min_length → lower limit
    do_sample=False → ensures deterministic output
    """
    summary = summarizer(
        text,
        max_length=80,
        min_length=30,
        do_sample=False
    )
    return summary[0]['summary_text']


# Step 2: Ask user for input text
print("Enter the text you want to summarize:")
user_text = input()

# Step 3: Generate summary
print("\n----- Summary -----\n")
print(generate_summary(user_text))
```
